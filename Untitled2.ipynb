```python
import kagglehub
import os
```

```python
path = kagglehub.dataset_download("kanchana1990/real-estate-data-london-2024")

print("Path to dataset files:", path)
```

```python
data_paths = [os.path.join(path, f) for pth, dirs, files in os.walk(path) for f in files]
print(data_paths)
```

```python
import pandas as pd
data = pd.read_csv(data_paths[0])
data
```

```python
# Based on the row "count", there are numbers of missing values in 4 out of the 9 columns.
# There are values with repetition on both "title" and "descriptionHtml" columns which should be checked to see if they are duplications or not.
# The data type of the column "price" is string, and for conveniece during data analysis, it will be converted to numerical format.
# The odd thing that caught my attention is that the highest value for both "bedrooms" and "bathrooms" is equal to 66, which needs to be analyzed in correlation with
# size and price.
# The values in the column "addedOn" will also be converted to datetime format.
```

```python
data['price'].sample(5)
```

```python
data['price'] = data['price'].str.replace('£', '')
data['price'] = data['price'].str.replace(',', '')
data['price'].astype(int)
```

```python
mask = data['price'] == 'POA'
data[mask]
```

```python
mask = data['propertyType'] == 'Plot'
data[mask]
```

```python
# Convrting to integer raised an error which is caused by the value "POA".
# Searching for the rows with this attribute shows that there is only one sample, which is also has 4
# missing values.
# For tackling this problem, the "POA" will be changed to Nane, and the column will be converted to integer type.
```

```python
data['price'] = pd.to_numeric(data['price'], errors='coerce')
data['price'].describe()
```

```python
date_added = pd.to_datetime(data['addedOn'], infer_datetime_format = True)
date_added
```

```python
# Univariate Analysis
```

```python
data['title'][0].find('for sale in ')
data['title'][0][16+len('for sale in ') : data['title'][0].find(', London')]
```

```python
loc = []
for i in range(len(data)):
  location = data.iloc[i]['title'][data.iloc[i]['title'].find('for sale in ') + len('for sale in') : ]
  loc.append(location)

loc = pd.DataFrame(loc, columns=['Location'])
print(loc)
```

```python
loc['Location'][0].split(',')
```

```python
import re
import random
pattern = r","
re.split(pattern, loc['Location'][0])
```

```python
postal_area = []
for i in range(len(loc)):
  postal_area.append(loc['Location'][i].split(',')[-1])

postal_area = pd.DataFrame(postal_area, columns=['Postal Area'])
print(postal_area)
```
